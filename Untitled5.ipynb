### Mapping Experiences to the Sphere

To map experiences defined by 'Emotional Valence' and 'Arousal' onto the sphere, we'll translate these two dimensions into spherical coordinates:

*   **Emotional Valence (ranges from -1 to 1)**: This will be mapped to the **azimuth angle** (`u_exp`), which determines the position around the 'equator' of the sphere. A linear mapping from -1 to 1 will span from `0` to `2 * pi` radians.
    *   `u_exp = (valence + 1) * pi`
    *   Negative valence (-1) maps to `0` radians (or `2*pi`).
    *   Neutral valence (0) maps to `pi` radians.
    *   Positive valence (1) maps to `2*pi` radians.

*   **Arousal (ranges from 0 to 1)**: This will be mapped to the **polar angle** (`v_exp`), which determines the vertical position from the 'north pole' to the 'south pole'. We will map low arousal to the equator and high arousal towards the north pole.
    *   `v_exp = (1 - arousal) * pi / 2`
    *   Low arousal (0) maps to `pi / 2` radians (the equator).
    *   High arousal (1) maps to `0` radians (the north pole).

Each experience will be represented as a point on the sphere with a constant radius (e.g., 1.0, matching the sphere's radius).

```python
import numpy as np
import plotly.graph_objects as go

# Prepare data for filtering
exp_data = np.array([[e['valence'], e['arousal']] for e in experiences])
names = [e['name'] for e in experiences]
colors = [e['color'] for e in experiences]

def get_coords(v_list, a_list):
    u = (np.array(v_list) + 1) * np.pi
    v = (1 - np.array(a_list)) * np.pi / 2
    x = 1.0 * np.sin(v) * np.cos(u)
    y = 1.0 * np.sin(v) * np.sin(u)
    z = 1.0 * np.cos(v)
    return x, y, z

fig = go.Figure()

# 1. Base Sphere
u_grid, v_grid = np.mgrid[0:2*np.pi:30j, 0:np.pi:30j]
sx, sy, sz = np.cos(u_grid)*np.sin(v_grid), np.sin(u_grid)*np.sin(v_grid), np.cos(v_grid)
fig.add_trace(go.Surface(x=sx, y=sy, z=sz, opacity=0.1, colorscale='Greys', showscale=False, hoverinfo='skip'))

# 2. Experience Points
x_e, y_e, z_e = get_coords(exp_data[:,0], exp_data[:,1])
fig.add_trace(go.Scatter3d(x=x_e, y=y_e, z=z_e, mode='markers+text',
                           marker=dict(size=8, color=colors, opacity=0.9),
                           text=names, name='All Experiences'))

# 3. Navigational Logic: Buttons for Filtering
buttons = []
filters = [
    ('Show All', lambda v, a: True),
    ('Positive Valence', lambda v, a: v > 0),
    ('Negative Valence', lambda v, a: v < 0),
    ('High Arousal', lambda v, a: a > 0.5),
    ('Calm States', lambda v, a: a < 0.3)
]

for label, logic in filters:
    mask = [logic(e['valence'], e['arousal']) for e in experiences]
    curr_x, curr_y, curr_z = get_coords(exp_data[mask, 0], exp_data[mask, 1])

    buttons.append(dict(
        method='update',
        label=label,
        args=[{'x': [sx, curr_x], 'y': [sy, curr_y], 'z': [sz, curr_z],
               'text': [None, np.array(names)[mask]]}]
    ))

fig.update_layout(
    updatemenus=[dict(buttons=buttons, direction='down', showactive=True, x=0.1, y=1.1, bgcolor='rgba(50,50,50,0.8)', font=dict(color='white'))],
    title='Ophanim Navigation: Multi-State Filtering',
    template='plotly_dark',
    scene=dict(xaxis_visible=False, yaxis_visible=False, zaxis_visible=False, aspectmode='cube')
)

fig.show()
```

## Define and Visualize the Experience Sphere

### Subtask:
Define the 'experience sphere' by its properties (e.g., center, radius) and generate points on its surface. Visualize this basic sphere using `matplotlib` to establish a foundation for later 'experience' mapping.

**Reasoning**:
To represent a sphere in 3D for visualization, I'll use spherical coordinates to generate a mesh of points on its surface. These points will then be plotted using `matplotlib`'s 3D plotting capabilities.

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define sphere parameters
radius = 1.0
center = np.array([0, 0, 0])

# Generate points on the surface of the sphere using spherical coordinates
u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(0, np.pi, 100)

x = radius * np.outer(np.cos(u), np.sin(v)) + center[0]
y = radius * np.outer(np.sin(u), np.sin(v)) + center[1]
z = radius * np.outer(np.ones(np.size(u)), np.cos(v)) + center[2]

# Create a 3D plot
fig = plt.figure(figsize=(8, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the sphere
ax.plot_surface(x, y, z, color='b', alpha=0.6)

# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('Basic Experience Sphere')

# Set equal aspect ratio
ax.set_box_aspect([1,1,1])

plt.show()

print("Basic 3D sphere generated and displayed.")
```
